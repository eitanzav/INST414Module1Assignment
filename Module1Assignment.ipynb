```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/eitanzavorin/Downloads/vgsales.csv')

df
```

```python
df.drop_duplicates(subset=['Name'], inplace=True)

df.drop(columns=['EU_Sales', 'NA_Sales', 'JP_Sales', 'Other_Sales', 'Platform'], inplace=True)

df
```

```python
fps_df = df.loc[df['Genre'] == 'Shooter']
fps_df
```

```python
top_10_games = fps_df.head(10)
print(top_10_games)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10_games['Name'], top_10_games['Global_Sales'], color='skyblue')
plt.xlabel('Game Name')
plt.ylabel('Global Sales (in millions)')
plt.title('Top 10 Highest Selling First-Person Shooter Games')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
publisher_sales = fps_df.groupby('Publisher')['Global_Sales'].sum()

# Select the top 5 publishers with the highest total sales
top_5_publishers = publisher_sales.nlargest(5)

# Plotting
plt.figure(figsize=(8, 6))
plt.pie(top_5_publishers, labels=top_5_publishers.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Top 5 Highest Selling Publishers (FPS ONLY)')
plt.show()
```
